### Exercício 1: Gráfico de Uma Função Linear

Crie um gráfico para a função y = 2x + 3 para valores de x no intervalo de 5 a 10.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2 * x + 3

# 500 quantidade de valores
x = np.linspace(5, 10, 500)
y = f(x)
# Cria o gráfico
plt.figure(figsize = (6, 4))
plt.plot(x, y, label = 'x,y')
# Adiciona um título e rótulos aos eixos
plt.title('Gráfico da Função: y = 2x + 3')
plt.xlabel('x')
plt.ylabel('y')
# Adiciona uma legenda
plt.legend()
# Mostra o gráfico
plt.grid(True)
plt.show()
```

### Exercício 6: Resolver uma Equação Quadrática

Escreva um script em Python que resolve a equação x^2 − 5x + 6 = 0 e imprime as raízes.

```python
import cmath
# cmath para numelhor complexos

def raizQuadratica(a,b,c):
    delta = cmath.sqrt(b**2 - 4*a*c)
    
    raiz = ((-1 * b) + delta) / (2 * a)
    raiz2 = ((-1 * b) - delta) / (2*a)
    print("Retorno x': "+str(raiz) +" x'': "+str(raiz2))


a, b, c = 1, -5, 6
raizQuadratica(a,b,c)
```

### Exercício 8: Equação Quadrática com Discriminante

Para a equação quadrática ax^2 + bx + c = 0, calcule o discriminante e determine o número de raízes reais.

```python
def raizQuadratica(a,b,c):
    delta = b**2 - 4*a*c
    
    if delta > 0:
        print("duas raizes reais e distintas.")
    elif delta==0:
        print("Uma raiz real dupla")
    else:
        print("nehuma raiz real")
        
    raiz = ((-1 * b) + cmath.sqrt(delta)) / (2 * a)
    raiz2 = ((-1 * b) - cmath.sqrt(delta)) / (2*a)
    print("Retorno x': "+str(raiz) +" x'': "+str(raiz2))
    
a = int(input ("Qual o valor de a: " ))
b = int(input ("Qual o valor de b: " ))
c = int(input ("Qual o valor de c: " ))

raizQuadratica(a,b,c)
```

### Exercício 9: Cálculo de Logaritmos

Escreva um script que calcula o logaritmo de base 10 de um número fornecido pelo usuário. Certifique-se de tratar o caso em que o logaritmo não pode ser definido (números negativos e zero).

```python
import math
a = float(input ("Qual o valor de a: " ))

if a > 0:
    print("valor: "+str(math.log10(a)))
else:
    print("logaritmo não definico para numero <= zero")
```
